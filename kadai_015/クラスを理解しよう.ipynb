```python
class Human:
    def __init__(self, name, age):
        # 属性（nameとage）を定義
        self.name = name
        self.age = age

    # 情報を出力するメソッド
    def printinfo(self):
        print(f"名前: {self.name}")
        print(f"年齢: {self.age}歳")


# Humanクラスのインスタンスを作成し、変数に代入
person = Human("たろう", 12)

# メソッドを呼び出して、情報を表示
person.printinfo()
```
