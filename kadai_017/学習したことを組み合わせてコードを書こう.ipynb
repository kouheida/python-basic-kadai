```python
# Humanクラスを定義
class Human:
    def __init__(self, name, age):
        self.name = name  # 名前を保存
        self.age = age    # 年齢を保存

    # 大人かどうかを判定・表示するメソッド
    def check_adult(self):
        if self.age >= 20:
            print(f"{self.name}（{self.age}歳）は大人です。")
        else:
            print(f"{self.name}（{self.age}歳）は大人ではありません。")


# 複数のHumanインスタンスを作成
person1 = Human("たろう", 18)
person2 = Human("はなこ", 22)
person3 = Human("ゆうた", 15)
person4 = Human("さくら", 30)

# リストに追加
people = [person1, person2, person3, person4]

# リストの中の全員に対してcheck_adultを実行
for person in people:
    person.check_adult()
```
